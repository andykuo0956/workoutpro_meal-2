### 好用的 Array 運算性質

```python
import numpy as np
# 向量相乘
b_1 = np.random.randint(1, high = 100, size=1000)
b_2 = np.random.randint(1, high= 100 , size=1000)
```

```python
# 向量元素相乘 - list
start_time = time.time()

for i in b_1:
  for j in b_2:
    k = i*j

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
# 向量元素相乘 - array
start_time = time.time()

print(np.multiply.outer(b_1,b_2))

print("--- %s seconds ---" % (time.time() - start_time))
```

## 建立 pd.DataFrame

### 建立pd.Series
- np.array
- dictionary

```python
# index
hot_pot_index = pd.Index(['老四川巴蜀麻辣燙','天母寧記麻辣火鍋','藍公館麻辣鍋'])
print(type(hot_pot_index))
```

```python
# array
hot_pot_rating = np.array([4.2,4.3,4.0])
hot_pot_reviews = np.array([3334,217,421])

print(type(hot_pot_rating))
print(type(hot_pot_reviews))
```

```python
#----------------------------
print('')
print('This is hot_pot_rating')
print(pd.Series(hot_pot_rating, index = hot_pot_index))
print('')
print('This is hot_pot_review')
print(pd.Series(hot_pot_reviews, index = hot_pot_index))
```

```python
# use dictionaty
hot_pot_rating_dict = {'老四川巴蜀麻辣燙': 4.2, '天母寧記麻辣火鍋': 4.3, '藍公館麻辣鍋': 4.0}
hot_pot_review_dict = {'老四川巴蜀麻辣燙': 3334, '天母寧記麻辣火鍋': 217, '藍公館麻辣鍋': 421}
hot_pot_rating_series = pd.Series(hot_pot_rating_dict)
hot_pot_review_series = pd.Series(hot_pot_review_dict)
print(type(hot_pot_rating_series))
#----------------------------
print('')
print('This is hot_pot_rating')
print(hot_pot_rating_series)
print('')
print('This is hot_pot_review')
print(hot_pot_review_series)
```

### 從 Series 到 DataFrame

```python
hot_pot_df = pd.DataFrame()
hot_pot_df['rating'] = hot_pot_rating_series
hot_pot_df['review'] = hot_pot_review_series
hot_pot_df
```

### 從 Dictionary 到 DataFrame

```python
pd.DataFrame([hot_pot_rating_dict, hot_pot_review_dict])
```
